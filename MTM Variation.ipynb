```python
import pdfplumber
```

```python
file_path_Sucden_Fw = r"C:\Users\qza\Downloads\DailyEquityReport_50327_20250526.pdf"
```

```python
ccy_fxrate = {}
ccy_variation = {}
valid_section_found = False

with pdfplumber.open(file_path_Sucden_Fw) as pdf:
    capture_section = None
    
    for page in pdf.pages:
        text = page.extract_text()
        lines = text.split('\n')
        
        for line in lines:
            if "Equity Summary at Close of Business" in line:
                capture_section = "Equity Summary Close"
                valid_section_found = True
                break
            elif "Equity Summary for Account 50327" in line:
                capture_section = "Equity Summary 50327"
                valid_section_found = True
                break

        if capture_section:
            for line in lines:
                parts = line.split()
                if len(parts) >= 6 and parts[0] != "Ccy":  
                    ccy = parts[0]
                    if capture_section == "Equity Summary Close":
                        variation = parts[2]
                        fxrate = parts[-1]
                    elif capture_section == "Equity Summary 50327":
                        variation = parts[4]
                        fxrate = parts[-1] 
                            
                    try:
                        variation_value = float(variation.replace(",", ""))
                        ccy_variation[ccy] = variation_value
                    except ValueError:
                        pass

                    # 提取 FxRate
                    try:
                        fxrate_value = float(fxrate.replace(",", ""))
                        ccy_fxrate[ccy] = fxrate_value
                    except ValueError:
                        pass

                if line.strip().startswith("ZAR"):
                    break

if not valid_section_found:
        print("No valid section found in the PDF.")
    
# Print results
print("Variation:")
for ccy, variation in ccy_variation.items():
    print(f"{ccy}: {variation}")

print("\nFxRate:")
for ccy, fxrate in ccy_fxrate.items():
    print(f"{ccy}: {fxrate}")
```
